## IMPORTING PACKAGES AND DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('matches.csv')
df
```

## EXPLORING DATA

```python
df[df['umpire3'].isnull()]
```

```python
df=df.drop(columns=['umpire3'])
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna({'umpire1':'Nitin Menon'},inplace=True)
df.isnull().sum()
```

```python
df.fillna({'umpire2':'Kumar Dharmasena'},inplace=True)
df.isnull().sum()
```

```python
df.fillna({'city':'Mumbai'},inplace=True)
df.isnull().sum()
```

```python
df[df['winner'].isnull()]
```

```python
df[df['player_of_match'].isnull()]
```

```python
df.fillna({'winner':'Unavailable','player_of_match':'Unavailable'},inplace=True)
df.isnull().sum()
```

## FINDING PLAYER WITH MOST NO.OF.PLAYER OF MATCH

```python
df['player_of_match'].value_counts().head(15)
```

```python
x=list(df['player_of_match'].value_counts().head(10).keys())
x
```

```python
y=list(df['player_of_match'].value_counts().head(10))
y
```

```python
plt.figure(figsize=(15,10))
plt.bar(x,y,color=['red','yellow','green','black','pink','blue','orange','violet','Indigo','Silver'])
plt.grid(True)
plt.title('Man of the Match')
plt.xlabel('Name of the Players')
plt.ylabel('Number of Awards')
plt.show()
```

## TEAMS WINNING BATTING FIRST

```python
batting_first=df[df['win_by_runs']!=0]
batting_first
```

```python
batting_first.head(20)
```

```python
plt.figure(figsize=(7,5))
sns.histplot(x = "win_by_runs",data = batting_first, kde = True,color='purple',bins=10)
plt.grid(True)
plt.title('Distribution of Win by Runs')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
list(batting_first['winner'].value_counts().keys())
```

```python
list(batting_first['winner'].value_counts())
```

```python
plt.figure(figsize=(7,7))
sns.countplot(y='winner',order=batting_first['winner'].value_counts().index[0:15],data=batting_first,palette = "Set2")
plt.title('Winning Teams After Batting First ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,12))
myexplode = [0.05, 0.05, 0.05, 0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%',explode = myexplode,shadow = True)
plt.show()
```
